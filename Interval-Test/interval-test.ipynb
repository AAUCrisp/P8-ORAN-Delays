```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
times = [0.0001,0.001,0.01,0.1,0.5,1,5,10]

means = []
mins = []
maxs = []

for i in range(1, 9):
    df = pd.read_csv(f"interval-tests/{i}.csv")
    mean = df["time"].mean()*1000
    mini = df["time"].min()*1000
    maxi = df["time"].max()*1000

    means.append(mean)
    mins.append(mini)
    maxs.append(maxi)
    
    
plt.plot(times, means, label="Mean")
plt.plot(times, mins, label="Min")
plt.plot(times, maxs, label="Max")
plt.xscale("log")
plt.grid()
plt.xlabel("Interval [ms]")
plt.ylabel("Latency [ms]")
plt.title("DL Latency vs Interval")
plt.legend()
plt.savefig("latency-vs-interval.png")
plt.show()
```
